```python
import numpy as np
import matplotlib.pyplot as plt
import random

from scipy import integrate
```

# function plot


```python
def funcplot(x,plot=False):
    func=0.5+np.sin(np.pi*x)
    if plot==True:   
        plt.figure(figsize=[10,8])
        plt.plot(x,func,'r-')
        plt.grid()
        plt.xlabel('x',fontsize=15)
        plt.ylabel('y',fontsize=15)
        plt.show()
    return func
```

```python
x=np.linspace(0.25,1.25,100)
```

```python
y=funcplot(x,plot=True)
```

# Simpson integration

```python
y=funcplot(x)
ans = integrate.simps(y, x)
```

```python
print('Area under the curve:',ans,'unit\u00b2')
```

# Monte Carlo integration

```python
def monteCarlo(func,uplim,lowlim,no_of_samples=2000,printval=True):
    N_array = np.random.uniform(lowlim,uplim,no_of_samples)
    funcsum=0
    for i in N_array:
        funcsum=funcsum+func(i)
    if printval==True:
        print('Integral using Monte Carlo algorithm:',(1.25-0.25)*funcsum/len(N_array))
    return (1.25-0.25)*funcsum/len(N_array)
```

```python
val=monteCarlo(funcplot,0.25,1.25,5000)
```

```python
y_intgral=[]
for i in np.arange(500,100500,1000):
    y_intgral.append(monteCarlo(funcplot,0.25,1.25,i,printval=False))
```

```python
plt.figure(figsize=[10,8])
plt.plot(np.arange(500,100500,1000),y_intgral,'r-')
plt.grid()
plt.xlabel('No of samples collected',fontsize=12)
plt.ylabel('Integrated value',fontsize=12)
plt.axhline(y=ans,color='blue',label='Simpson integration')
plt.legend()
plt.show()
```
