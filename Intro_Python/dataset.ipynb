```python
#Exercise for exploring large dataset


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Stars.csv', header=0)
print(df)
```

```python
df['Type'] #for example we want to check pandas read the data correctly
```

```python
df.groupby('Type').count().plot.bar()
```

```python
df.columns
```

```python
#Plot all the data from star.csv. We created four subplots for temperature, L, R and A_M

fig,axs = plt.subplots(2,2, figsize=(15,15))
df.Temperature.plot(kind='line',color='b', label = 'Temperature', linewidth=1.5, alpha=0.8, linestyle='-', ax=axs[0,0])
df.L.plot(kind='line',color='r', label = 'L', linewidth=1.5, alpha=0.8, linestyle='-', ax=axs[0,1])
df.R.plot(kind='line',color='g', label = 'R', linewidth=1.5, alpha=0.8, linestyle='-', ax=axs[1,0])
df.A_M.plot(kind='line',color='y', label = 'A_M', linewidth=1.5, alpha=0.8, linestyle='-', ax=axs[1,1])

#include legend according to the subplot location
#for coordinate [0,0]
axs[0,0].legend(loc='upper right')
axs[0,0].set_xlabel('Index')
axs[0,0].set_ylabel('Temperature')

#for coordinate [0,1]
axs[0,1].legend(loc='upper right')
axs[0,1].set_xlabel('Index')
axs[0,1].set_ylabel('L')

#for coordinate [1,0]
axs[1,0].legend(loc='upper right')
axs[1,0].set_xlabel('Index')
axs[1,0].set_ylabel('R')

#for coordinate [1,1]
axs[1,1].legend(loc='upper right')
axs[1,1].set_xlabel('Index')
axs[1,1].set_ylabel('A_M')
```

```python
#For scatter plot where x = Type and y = L
df.plot(kind='scatter',x='Type', y= 'A_M',color = 'cyan')
plt.xlabel('Type')
plt.ylabel('A_M')
plt.title('Type vs A_M')
```

We can see from the scatter plot above the relation with A_M and Type, when the type of number of star increases, the A_M decreases
